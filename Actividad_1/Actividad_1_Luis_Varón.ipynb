```python
# Ejercicio 1: Clientes Duplicados y Segmentación

clientes = ['Ana', 'Luis', 'Pedro', 'Ana', 'Marta', 'Luis']
clientes_activos = {'Ana', 'Luis', 'Marta'}
clientes_inactivos = {'Pedro', 'Luis', 'Juan'}
```

```python
# Eliminar duplicados

clientes_unicos = set(clientes) # Conversión a set()

print(clientes_unicos)
```

```python
# Idenmtificar activos e inactivos

clientes_activos = clientes_unicos.intersection(clientes_activos) # Intersección para tener clientes activos
clientes_inactivos = clientes_unicos.intersection(clientes_inactivos) # Intersección para tener clientes inactivos

print(clientes_activos)
print(clientes_inactivos)
```

```python
# Solo clientes activos

clientes_activos = clientes_unicos.difference(clientes_inactivos) # Clientes activos filtrando con difference

print(clientes_activos)
```

```python
# Activos o VIP y pendientes

for cliente in clientes_unicos: # Iterar cada cliente en la lista clientes_unicos
    if cliente in clientes_activos: # Comprobar si está en activos
        print(f'{cliente} es un cliente VIP') # Imprimir mensaje
    else:
        print(f'{cliente} está Pendiente') # Imprimir mensaje
```

```python
# Ejercicio 2: Productos y Análisis de Inventario
ventas = {'Laptop', 'Mouse', 'Monitor'}
inventario = {'Laptop', 'Teclado', 'Mouse', 'Cámara'}
```

```python
# Catálogo total

catalogo_total = ventas.union(inventario) # Usando union() para tener el catalogo total

print(catalogo_total)
```

```python
# Productos vendidos en inventario

productos_vendidos = ventas & inventario # Usando &

print(productos_vendidos)
```

```python
# Productos sin ventas

productos_sin_ventas = inventario - ventas # Usando -

print(productos_sin_ventas)
```

```python
# Revisión de stock

lista_inventario = list(inventario) # Convertir a lista con list()
i = 0 # Variable iterable

while i < len(lista_inventario):
  print(lista_inventario[i], "Revisión de stock") # Imprimir cada producto con el iterable
  i += 1 # Simular iteración para evitar bucle infinito

```

```python
# Ejercicio 3: Estudiantes y Cursos

curso_python = {'Carlos', 'Lucía', 'Felipe', 'Elena'}
curso_sql = {'Elena', 'Ricardo', 'Lucía', 'Andrés'}
```

```python
# Estudiantes inscritos en ambos cursos
ambos_cursos = curso_python & curso_sql # Intersección usando &

print(ambos_cursos)
```

```python
# Estudiantes solo en curso de python
solo_python = curso_python - curso_sql # Diferencia usando -

print(solo_python)
```

```python
# Todos los estudiantes
todos = curso_python.union(curso_sql) # Unión usando union()

print(todos)
```

```python
# Identificar doble titulación
for estudiante in todos:
  if estudiante in curso_python and estudiante in curso_sql: # Puede ser condicional así o anidado, lo prefiero así
    print(f'{estudiante} tiene doble titulación')
  else: # De esa forma con else ya estoy tomando los "restantes"
    print(f'{estudiante} solo tiene una titulación')
```
