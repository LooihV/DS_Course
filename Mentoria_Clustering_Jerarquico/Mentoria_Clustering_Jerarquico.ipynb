```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
data = {
    'Persona': ['Ana', 'Luis', 'Carlos', 'Sofia', 'Mateo', 'Laura'],
    'Ejercicio': [1,2,7,8,15,16],
    'Agua':[1,2,6,7,10,11]
}

df = pd.DataFrame(data)
df.head()
```

```python
x = df[['Ejercicio','Agua']]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Ejercicio'],df['Agua'])
for i, persona in enumerate(df['Persona']):
    plt.text(
        df['Ejercicio'][i]+0.2,
        df['Agua'][i],
        persona
    )
plt.xlabel('Horas de ejercicio')
plt.ylabel('Consumo de agua')
plt.title('Diagrama de dispersión')
plt.show()
```

```python
plt.figure(figsize=(7,5))
linkage_data = linkage(x, method='ward')
dendrogram(linkage_data, labels=df['Persona'].values)
plt.xlabel('Personas')
plt.ylabel('Distancia')
plt.title('Dendrograma')
plt.show()
```

```python
modelo = AgglomerativeClustering(n_clusters=3)
clusters = modelo.fit_predict(x)
df['Cluster'] = clusters
df
```
